```python
def plot_age(age):
    plt.close()
    sns.set_style('ticks', {'axes.spines.right': False, 'axes.spines.top': False, 'xtick.bottom': False})
    ax = subplot(111)
    bins = np.linspace(6.5, 18.5, 13)
    print('Mean and median age:' , round(np.mean(age),2), np.median(age), '\n')
    #print(len(age[np.where(age<8)]))
    median = '%.0f' % np.median(age)
    values, base = np.histogram(age, bins=bins)
    percent = 100*(values/participants)
    plt.hist(age, bins = bins, fc = 'none', lw = 2.5, ec = sns.xkcd_rgb['cerulean blue'])
    for i  in range(len(values)):
        if values[i]>0:
            pcstr = '%.0f' % percent[i]
            plt.text(base[i+1]-0.5, values[i]+(0.05*np.max(values)), pcstr+' %', color='k', fontweight='bold', va='center', ha = 'center', fontsize = 11)
    plt.text(7, 0.98*np.max(values), 'Participant Total: '+str(participants), color='k', fontweight='bold', va='center', ha = 'left', fontsize = 14)
    plt.text(7, 0.9*np.max(values), 'Median age: '+median+' yrs', color='k', fontweight='bold', va='center', ha = 'left', fontsize = 15)
    plt.axes().yaxis.set_minor_locator(MultipleLocator(5))
    plt.xlabel('Age of participant (years)', fontsize = 20)
    plt.ylabel('Number of participants', fontsize = 20)
    plt.xlim(6.4,18.6)
    ax.xaxis.set_major_locator(MaxNLocator(14))
    plt.ylim(0,1.1*np.max(values))
    plt.savefig(plotpath+'AgeDistribution.png', bbox_inches='tight')
    plt.savefig(plotpath+'AgeDistribution.pdf', bbox_inches='tight')
```

```python
def stackedbarplot(combinedanswers, options, labels, colours, figurename, numbercolumns = 3):
    plt.close()
    sns.set_style('ticks', {'axes.spines.right': False, 'axes.spines.top': False, 'axes.spines.left': False, 'ytick.left': False})
    ind = [0,0.85,2]
    start = 0
    ax = subplot(111)
    num = 1/len(options)
    for i in range(len(options)):
        option = combinedanswers[:,i]
        plt.barh(ind, option, left = start, color = sns.xkcd_rgb['cerulean blue'], label = legendlabs[i], alpha = 1-num*i)
        #plt.barh(ind, option, left = start, color = cols[i], label = legendlabs[i])
        start = start+option
    plt.xlim(0,1)
    plt.yticks(ind, ('Male', 'Female', 'All'), fontsize = 18)
    plt.xticks(fontsize = 18)
    ax.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1))
    plt.legend(bbox_to_anchor=(0, 0.99, 1, .05), loc=3, ncol=numbercolumns, borderaxespad=0., fontsize = 15)
    plt.minorticks_on()
    fig = plt.gcf()
    fig.set_size_inches(8.7,6)
    plt.savefig(plotpath+'bar_'+figurename+'.pdf', bbox_inches='tight')
    #plt.savefig(plotpath+'bar_'+figurename+'.png', bbox_inches='tight')
    #plt.show()
    sns.set()
```

```python
def label_ratingsubplot(appname, rating, values):
    plt.text(1, np.max(values)*0.9, appname, weight = 'bold')
    plt.text(1, np.max(values)*0.8, 'Mean: %1.1f/5' %rating)
```

```python
def plot_pie(labels, sizes, colors, explode, title = 'title', filename = 'filename'):
    plt.close()
    #colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
    plt.pie(sizes, labels=labels, colors=colors,
            autopct='%1.1f%%', shadow=True, startangle=140, radius = 0.8, textprops=dict(fontsize = 15))
    plt.axis('equal')
    plt.savefig(plotpath+filename+'.pdf', bbox_inches='tight')
    plt.title(title, fontsize = 15)
    plt.savefig(plotpath+filename+'.png', bbox_inches='tight')
```

```python
plt.close()
def histrating(usedratedapp, rating, appname):
    bins = np.linspace(-0.5, 5.5, 7)
    values, base = np.histogram(usedratedapp, bins=bins)
    percent = 100*(values/len(usedratedapp))
    plt.hist(usedratedapp, bins = bins, fc='none', lw = 2.5, ec = sns.xkcd_rgb['cerulean blue'])
    plt.ylim(0, np.max(values)*1.1)
    plt.xlim(0.4, 5.6)
    label_ratingsubplot(appname, rating, values)
    ax.yaxis.set_major_locator(MaxNLocator(5))
    ax.xaxis.set_ticks_position('bottom')
    
fig = plt.figure(figsize=(8,10))
gs1 = gridspec.GridSpec(4, 2)#, height_ratios=[1,1,0.85])
gs1.update(left=0.11, right=0.99, bottom = 0.11, top = 0.99, wspace=0.10, hspace = 0)
s, mb, ma, mp, kc, ms, hm, tq = np.array(postdf['s']), np.array(postdf['mb']), np.array(postdf['ma']), np.array(postdf['mp']), np.array(postdf['kc']), np.array(postdf['ms']), np.array(postdf['hm']), np.array(postdf['tq'])
apps = s, mb, ma, mp, kc, ms, hm, tq
#apps = [np.array(postdf['s'], postdf['mb'], postdf['ma'], postdf['mp'], 
#        postdf['kc'], postdf['ms'], postdf['hm'], postdf['tq']]
ratings, appusernums, totalappusers = [], [], []
for app in apps:
    appfloat = app.astype(float)
    usedapp =  (appfloat[np.where(appfloat != 99)])
    usedratedapp =  (usedapp[np.where(usedapp != 66)])
    totalusers = len(usedapp)
    appusers =  len(usedratedapp)
    rating = np.mean(usedratedapp)
    ratings = np.append(ratings, rating)
    totalappusers = np.append(totalappusers, totalusers)
    appusernums = np.append(appusernums, appusers)
    sns.set_style('ticks', {'axes.spines.right': False, 'axes.spines.top': False})
    if app is s:
        appname = 'Scratch'
        ax = plt.subplot(gs1[0,0])
        histrating(usedratedapp, rating, appname)
        setp(ax.get_xticklabels(), visible=False)
    elif app is mb: 
        appname = 'Micro:bit'
        ax =  plt.subplot(gs1[0,1])
        histrating(usedratedapp, rating, appname)
        setp(ax.get_xticklabels(), visible=False)
    elif app is ma:
        appname = 'Make Art'
        ax =  plt.subplot(gs1[1,0])
        histrating(usedratedapp, rating, appname)
        setp(ax.get_xticklabels(), visible=False)
        plt.ylabel('Number of participants', fontsize = 20)
    elif app is mp:
        appname = 'Make Pong'
        ax =  plt.subplot(gs1[1,1])
        histrating(usedratedapp, rating, appname)
        setp(ax.get_xticklabels(), visible=False)
    elif app is kc:
        appname = 'Kano Code'
        ax =  plt.subplot(gs1[2,0])
        histrating(usedratedapp, rating, appname)
        setp(ax.get_xticklabels(), visible=False)
    elif app is ms:
        appname = 'Make Snake'
        ax1 =  plt.subplot(gs1[2,1])
        histrating(usedratedapp, rating, appname)
        ax1.set_xticks([1,2,3,4,5])
        plt.xlabel('Rating /5', fontsize = 20)
    elif app is hm:
        appname = 'Hack Minecraft'
        ax1 =  plt.subplot(gs1[3,0])
        histrating(usedratedapp, rating, appname)
        ax1.set_xticks([1,2,3,4,5])
        plt.xlabel('Rating /5', fontsize = 20)
    elif app is tq:
        appname = 'Terminal Quest'
        ax1 =  plt.subplot(gs1[3,1])
        histrating(usedratedapp, rating, appname)
        ax1.set_xticks([1,2,3,4,5])
        plt.xlabel('Rating /5', fontsize = 20)
    print('\n', appname, appusers) 

plt.savefig(plotpath+'AppRatings.png', bbox_inches='tight')
plt.savefig(plotpath+'AppRatings.pdf', bbox_inches='tight')
```
