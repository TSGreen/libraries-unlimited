```python
def plot_age(age):
    plt.close()
    sns.set_style('ticks', {'axes.spines.right': False, 'axes.spines.top': False, 'xtick.bottom': False})
    ax = subplot(111)
    bins = np.linspace(6.5, 18.5, 13)
    print('Mean and median age:' , round(np.mean(age),2), np.median(age), '\n')
    #print(len(age[np.where(age<8)]))
    median = '%.0f' % np.median(age)
    values, base = np.histogram(age, bins=bins)
    percent = 100*(values/participants)
    plt.hist(age, bins = bins, fc = 'none', lw = 2.5, ec = sns.xkcd_rgb['cerulean blue'])
    for i  in range(len(values)):
        if values[i]>0:
            pcstr = '%.0f' % percent[i]
            plt.text(base[i+1]-0.5, values[i]+(0.05*np.max(values)), pcstr+' %', color='k', fontweight='bold', va='center', ha = 'center', fontsize = 11)
    plt.text(7, 0.98*np.max(values), 'Participant Total: '+str(participants), color='k', fontweight='bold', va='center', ha = 'left', fontsize = 12)
    plt.text(7, 0.9*np.max(values), 'Median age: '+median+' yrs', color='k', fontweight='bold', va='center', ha = 'left', fontsize = 12)
    #plt.axes().yaxis.set_minor_locator(MultipleLocator(5))
    plt.xlabel('Age of participant (years)', fontsize = 20)
    plt.ylabel('Number of participants', fontsize = 20)
    plt.xlim(6.4,18.6)
    ax.xaxis.set_major_locator(MaxNLocator(14))
    plt.ylim(0,1.1*np.max(values))
    plt.savefig(plotpath+'AgeDistribution.png', bbox_inches='tight')
    plt.savefig(plotpath+'AgeDistribution.pdf', bbox_inches='tight')
```

```python
def stackedbarplot(combinedanswers, options, labels, colours, figurename, numbercolumns = 3, samplesize = 0):
    plt.close()
    sns.set_style('ticks', {'axes.spines.right': False, 'axes.spines.top': False, 'axes.spines.left': False, 'ytick.left': False})
    ind = [0,0.85,2]
    start, pos, = 0, [0,0,0]
    ax = subplot(111)
    num = 1/len(options)
    for i in range(len(options)):
        option = combinedanswers[:,i]
        plt.barh(ind, option, left = start, label = legendlabs[i])
        for k in range(len(ind)):
            xpos = pos[k]+option[k]/2
            percent = int(option[k]*100)
            #print(i, k, percent, pos, xpos, ind[k])
            if percent >= 10:
                plt.annotate(str(percent)+' %', xy=(xpos,ind[k]), ha = 'center', fontsize = 15, color = '1')
            elif percent<3:
                pass
            else: 
                plt.annotate(str(percent), xy=(xpos,ind[k]), ha = 'center', fontsize = 15, color = '1')
        start = start+option
        pos = start
    plt.xlim(0,1)
    plt.yticks(ind, ('Male', 'Female', 'All'), fontsize = 18)
    plt.xticks(fontsize = 18)
    ax.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1))
    plt.legend(bbox_to_anchor=(0, 0.99, 1, .05), loc=3, ncol=numbercolumns, borderaxespad=0., fontsize = 15)
    plt.minorticks_on()
    fig = plt.gcf()
    plt.figtext(0.9,0.135, ('Based on sample of %i participants' %samplesize), fontsize = 10, ha = 'right')
    fig.set_size_inches(8.7,6)
    plt.savefig(plotpath+'bar_'+figurename+'.pdf', bbox_inches='tight')
    #plt.savefig(plotpath+'bar_'+figurename+'.png', bbox_inches='tight', dpi=600)
    #plt.show()
    sns.set()
```

```python
def prepoststackedbarplot(combinedanswers, options, labels, colours, figurename, numbercolumns = 3, samplesize=0):
    plt.close()
    sns.set_style('ticks', {'axes.spines.right': False, 'axes.spines.top': False, 'axes.spines.left': False, 'ytick.left': False})
    ind = [0,1]
    ax = subplot(111)
    num = 1/len(options)
    start, pos, = 0, [0,0,0]
    for i in range(len(options)):
        option = combinedanswers[:,i]
        plt.barh(ind, option, left = start, label = legendlabs[i])
        for k in range(len(ind)):
            xpos = pos[k]+option[k]/2
            percent = int(option[k]*100)
            #print(i, k, percent, pos, xpos, ind[k])
            if percent >= 10:
                plt.annotate(str(percent)+' %', xy=(xpos,ind[k]), ha = 'center', fontsize = 15, color = '1')
            elif percent<3:
                pass
            else: 
                plt.annotate(str(percent), xy=(xpos,ind[k]), ha = 'center', fontsize = 15, color = '1')
        start = start+option
        pos = start
    plt.xlim(0,1)
    plt.yticks(ind, ('Post', 'Pre'), fontsize = 18)
    plt.xticks(fontsize = 18)
    ax.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1))
    plt.legend(bbox_to_anchor=(0, 0.99, 1, .05), loc=3, ncol=numbercolumns, borderaxespad=0., fontsize = 15)
    plt.minorticks_on()
    fig = plt.gcf()
    fig.set_size_inches(8.7,6)
    plt.figtext(0.9,0.135, ('Based on sample of %i participants' %samplesize), fontsize = 10, ha = 'right')
    plt.savefig(plotpath+'bar_'+figurename+'.pdf', bbox_inches='tight')
    #plt.savefig(plotpath+'bar_'+figurename+'.png', bbox_inches='tight', dpi=600)
    #plt.show()
    sns.set()
```

```python
def prestackedbarplot(combinedanswers, options, labels, colours, figurename, numbercolumns = 3):
    plt.close()
    sns.set_style('ticks', {'axes.spines.right': False, 'axes.spines.top': False, 'axes.spines.left': False, 'ytick.left': False})
    ind = [0]
    start = 0
    ax = subplot(111)
    num = 1/len(options)
    for i in range(len(options)):
        option = combinedanswers[:,i]
        #plt.barh(ind, option, left = start, color = sns.xkcd_rgb['cerulean blue'], label = legendlabs[i], alpha = 1-num*i)
        plt.barh(ind, option, left = start, label = legendlabs[i])
        start = start+option
    plt.xlim(0,1)
    plt.yticks(ind, ('Pre'), fontsize = 18)
    plt.xticks(fontsize = 18)
    ax.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1))
    plt.legend(bbox_to_anchor=(0, 0.99, 1, .05), loc=3, ncol=numbercolumns, borderaxespad=0., fontsize = 15)
    plt.minorticks_on()
    fig = plt.gcf()
    fig.set_size_inches(8.7,6)
    plt.savefig(plotpath+'bar_'+figurename+'.pdf', bbox_inches='tight')
    plt.savefig(plotpath+'bar_'+figurename+'.png', bbox_inches='tight', dpi=600)
    #plt.show()
    sns.set()
```

```python
def label_ratingsubplot(appname, rating, values):
    plt.text(1, np.max(values)*0.9, appname, weight = 'bold')
    plt.text(1, np.max(values)*0.8, 'Mean: %1.1f/5' %rating)
```

```python
def plot_pie(labels, sizes, colors, explode, title = 'title', filename = 'filename'):
    plt.close()
    #colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
    plt.pie(sizes, labels=labels, colors=colors,
            autopct='%1.1f%%', shadow=True, startangle=140, radius = 0.8, textprops=dict(fontsize = 15))
    plt.axis('equal')
    plt.savefig(plotpath+filename+'.pdf', bbox_inches='tight')
    plt.title(title, fontsize = 15)
    plt.savefig(plotpath+filename+'.png', bbox_inches='tight')
```

```python
def add_workshop(gender, age, priorcoding, s, mb, ma, mp, kc, ms, bk, hm, tq, kanoex, challenge, learnmore, libraryvisitbefore, libraryvisit,  itaccess, usecomp, usedevices, location):
    pretqlocs = ['dhaka_1', 'dhaka_2', 'coxsbazar', 'bandarban', 'rangamati', 'dhaka_3', 'jessore', 'khulna', 'narayanganj', 'nilphamari', 'thakurgaon', 'panchagarh']
    if location == 'munshiganj':
        genderR, ageR, priorcodingR, sR, mbR, maR, mpR, kcR, msR, bkR, hmR, learnmoreR, libraryvisitR, usedevicesR = np.loadtxt('feedback_data_munshiganj.txt', unpack=True, dtype = str)
        kanoexR, challengeR, libraryvisitbeforeR,  itaccessR, usecompR, tqR  = [], [], [], [], [], []
        kanoexR, challengeR, libraryvisitbeforeR,  itaccessR, usecompR, tqR  = np.zeros(len(genderR))+5000,np.zeros(len(genderR))+5000, np.zeros(len(genderR))+5000,np.zeros(len(genderR))+5000,np.zeros(len(genderR))+5000, np.zeros(len(genderR))+99
    elif location in pretqlocs:
        genderR, ageR, priorcodingR, sR, mbR, maR, mpR, kcR, msR, bkR, hmR, kanoexR, challengeR, learnmoreR, libraryvisitbeforeR, libraryvisitR,  itaccessR, usecompR, usedevicesR = np.loadtxt('feedback_data_'+location+'.txt', unpack=True, dtype = str)
        tqR = np.zeros(len(genderR))+99
    else:
        genderR, ageR, priorcodingR, sR, mbR, maR, mpR, kcR, msR, bkR, hmR, tqR, kanoexR, challengeR, learnmoreR, libraryvisitbeforeR, libraryvisitR,  itaccessR, usecompR, usedevicesR = np.loadtxt('feedback_data_'+location+'.txt', unpack=True, dtype = str)

    gender = np.concatenate((gender, genderR))
    age = np.concatenate((age, ageR))
    priorcoding = np.concatenate((priorcoding, priorcodingR))
    s = np.concatenate((s, sR))
    mb = np.concatenate((mb, mbR))
    ma = np.concatenate((ma, maR))
    mp = np.concatenate((mp, mpR))
    kc = np.concatenate((kc, kcR))
    ms = np.concatenate((ms, msR))
    bk = np.concatenate((bk, bkR))
    hm = np.concatenate((hm, hmR))
    tq = np.concatenate((tq, tqR))
    learnmore = np.concatenate((learnmore, learnmoreR))
    libraryvisit = np.concatenate((libraryvisit, libraryvisitR))
    usedevices = np.concatenate((usedevices, usedevicesR))
    kanoex = np.concatenate((kanoex, kanoexR))
    challenge = np.concatenate((challenge, challengeR))
    libraryvisitbefore = np.concatenate((libraryvisitbefore, libraryvisitbeforeR))
    itaccess = np.concatenate((itaccess, itaccessR))
    usecomp = np.concatenate((usecomp, usecompR))

        
    return(gender, age, priorcoding, s, mb, ma, mp, kc, ms, bk, hm, tq, kanoex, challenge, 
           learnmore, libraryvisitbefore, libraryvisit,  itaccess, usecomp, usedevices)
```

```python
def convertnumbers(x):
    if x <3:
        x = 'notatall'
    elif x<5:
        x = 'notmuch'
    elif x<7:
        x ='unsure'
    elif x<9:
        x = 'somewhat'
    else:
        x = 'very'
    return x
```
