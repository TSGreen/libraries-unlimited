```python
import numpy as np
import pandas as pd
```

```python
colnames = ('Gender, Age, Pre_CanDoCoding, Pre_CodingImportance, Pre_CodingInterest, ITaccess, UseComputersOften, '
            'VisitLibraryBefore, DesireLibraryVisit, Post_CanDoCoding, Post_CodingInterest, Post_CodingImportance, WorkshopFeedback, ' 
            'Scratch_rating, Microbit_rating, MakeArt_rating, MakePong_rating, KanoCode_rating, MakeSnake_rating, HackMinecraft_rating, TerminalQuest_rating, WillVisitLibrary, WillUseDevices').split(', ')
districts = ('pabna, meherpur, kushtia, rajshahi, nawabganj, natore').split(', ')
k = 0
for district in districts:
    df = pd.read_csv('feedback_complete_'+district+'.txt', sep="\t", names=colnames)
    df['Pre_CanDoCoding'].replace({'y':'yes', 'n':'no', 'dk':'dontknow'}, inplace = True)
    df['Post_CanDoCoding'].replace({'y':'yes', 'n':'no', 'dk':'dontknow'}, inplace = True)
    columns = ('Scratch_rating, Microbit_rating, MakeArt_rating, MakePong_rating, KanoCode_rating, MakeSnake_rating, HackMinecraft_rating, TerminalQuest_rating').split(', ')
    for col in columns:
        df[col].replace({66:'not_rated', 99:'not_used'}, inplace = True)
    df.to_csv('WorkshopData/'+district+'.csv')
```

```python
#colnames = ('gender, age, candocoding, codingimport, codingint, itaccess, usecomp, libraryvisitbefore, libvisitdesire').split(', ')
def convertnumbers(x):
    if x <3:
        x = 'notatall'
    elif x<5:
        x = 'notmuch'
    elif x<7:
        x ='unsure'
    elif x<9:
        x = 'somewhat'
    else:
        x = 'very'
    return x

colnames = ('Gender, Age, Pre_CanDoCoding, Pre_CodingImportance, Pre_CodingInterest, ITaccess, UseComputersOften, '
            'VisitLibraryBefore, DesireLibraryVisit').split(', ')
districts = ('patuakhali, barguna, barishal').split(', ')
for district in districts:
    df = pd.read_csv('feedback_predata_'+district+'.txt', sep="\t ", names=colnames)
    df['Pre_CanDoCoding'].replace({'y':'yes', 'n':'no', 'dk':'dontknow'}, inplace = True)
    df['Pre_CodingImportance'] = df['Pre_CodingImportance'].apply(convertnumbers)
    df['Pre_CodingInterest'] = df['Pre_CodingInterest'].apply(convertnumbers)
    df['DesireLibraryVisit'] = df['DesireLibraryVisit'].apply(convertnumbers)
    df.to_csv('WorkshopData/'+district+'_PreWorkshop.csv')
```

```python
def add_workshop(gender, age, priorcoding, s, mb, ma, mp, kc, ms, bk, hm, tq, kanoex, challenge, learnmore, libraryvisitbefore, libraryvisit,  itaccess, usecomp, usedevices, location):
    pretqlocs = ['dhaka_1', 'dhaka_2', 'coxsbazar', 'bandarban', 'rangamati', 'dhaka_3', 'jessore', 'khulna', 'narayanganj', 'nilphamari', 'thakurgaon', 'panchagarh']
    if location == 'munshiganj':
        genderR, ageR, priorcodingR, sR, mbR, maR, mpR, kcR, msR, bkR, hmR, learnmoreR, libraryvisitR, usedevicesR = np.loadtxt('feedback_data_munshiganj.txt', unpack=True, dtype = str)
        kanoexR, challengeR, libraryvisitbeforeR,  itaccessR, usecompR, tqR  = [], [], [], [], [], []
        kanoexR, challengeR, libraryvisitbeforeR,  itaccessR, usecompR, tqR  = np.zeros(len(genderR))+5000,np.zeros(len(genderR))+5000, np.zeros(len(genderR))+5000,np.zeros(len(genderR))+5000,np.zeros(len(genderR))+5000, np.zeros(len(genderR))+99
    elif location in pretqlocs:
        genderR, ageR, priorcodingR, sR, mbR, maR, mpR, kcR, msR, bkR, hmR, kanoexR, challengeR, learnmoreR, libraryvisitbeforeR, libraryvisitR,  itaccessR, usecompR, usedevicesR = np.loadtxt('feedback_data_'+location+'.txt', unpack=True, dtype = str)
        tqR = np.zeros(len(genderR))+99
    else:
        genderR, ageR, priorcodingR, sR, mbR, maR, mpR, kcR, msR, bkR, hmR, tqR, kanoexR, challengeR, learnmoreR, libraryvisitbeforeR, libraryvisitR,  itaccessR, usecompR, usedevicesR = np.loadtxt('feedback_data_'+location+'.txt', unpack=True, dtype = str)

    gender = np.concatenate((gender, genderR))
    age = np.concatenate((age, ageR))
    priorcoding = np.concatenate((priorcoding, priorcodingR))
    s = np.concatenate((s, sR))
    mb = np.concatenate((mb, mbR))
    ma = np.concatenate((ma, maR))
    mp = np.concatenate((mp, mpR))
    kc = np.concatenate((kc, kcR))
    ms = np.concatenate((ms, msR))
    bk = np.concatenate((bk, bkR))
    hm = np.concatenate((hm, hmR))
    tq = np.concatenate((tq, tqR))
    learnmore = np.concatenate((learnmore, learnmoreR))
    libraryvisit = np.concatenate((libraryvisit, libraryvisitR))
    usedevices = np.concatenate((usedevices, usedevicesR))
    kanoex = np.concatenate((kanoex, kanoexR))
    challenge = np.concatenate((challenge, challengeR))
    libraryvisitbefore = np.concatenate((libraryvisitbefore, libraryvisitbeforeR))
    itaccess = np.concatenate((itaccess, itaccessR))
    usecomp = np.concatenate((usecomp, usecompR))

        
    return(gender, age, priorcoding, s, mb, ma, mp, kc, ms, bk, hm, tq, kanoex, challenge, 
           learnmore, libraryvisitbefore, libraryvisit,  itaccess, usecomp, usedevices)
```

```python
readgender, readage, priorcoding, reads, readmb, readma, readmp, readkc, readms, readbk, readhm, readtq, kanoex, challenge, learnmore, libraryvisitbefore, libraryvisit,  itaccess, usecomp, usedevices = [], [], [], [], [],[], [], [], [], [],[], [], [], [], [],[], [], [], [], []


locations = ['munshiganj', 'dhaka_1', 'dhaka_2', 'coxsbazar', 'bandarban', 'rangamati', 'dhaka_3', 'jessore', 'khulna', 'narayanganj', 'nilphamari', 'thakurgaon', 'panchagarh', 'habiganj', 'sylhet', 'rangpur', 'lalmonirhat', 'dinajpur', 'shatkhira', 'bagerhat', 'mymensingh', 'sherpur', 'jamalpur', 'gazipur', 'tangail', 'manikganj']
#locations =['munshiganj']
k=0
app_columns = ('BuildKano_rating, Scratch_rating, Microbit_rating, MakeArt_rating, MakePong_rating, KanoCode_rating, MakeSnake_rating, HackMinecraft_rating, TerminalQuest_rating').split(', ')
colnames = ('Gender, Age, CodedBefore, Scratch_rating, Microbit_rating, MakeArt_rating, MakePong_rating, KanoCode_rating, MakeSnake_rating, BuildKano_rating, HackMinecraft_rating, TerminalQuest_rating, KanoDeviceFeedback, ChallengingAspect, Want_to_LearnMoreCoding, VisitLibraryBefore, WillVisitLibrary, ITaccess, UseComputersOften, WillUseDevices').split(', ')

for liblocation in locations:
    readgender, readage, priorcoding, reads, readmb, readma, readmp, readkc, readms, readbk, readhm, readtq, kanoex, challenge, learnmore, libraryvisitbefore, libraryvisit,  itaccess, usecomp, usedevices = [], [], [], [], [],[], [], [], [], [],[], [], [], [], [],[], [], [], [], []
    gender, age, priorcoding, s, mb, ma, mp, kc, ms, bk, hm, tq, kanoex, challenge, learnmore, libraryvisitbefore, libraryvisit,  itaccess, usecomp, usedevices = add_workshop(readgender, readage, priorcoding, reads, readmb, readma, readmp, readkc, readms, readbk, readhm, readtq, kanoex, challenge, learnmore, libraryvisitbefore, libraryvisit,  itaccess, usecomp, usedevices, location = liblocation)
    newdf = pd.DataFrame(zip(gender, age, priorcoding, s, mb, ma, mp, kc, ms, bk, hm, tq, kanoex, challenge, learnmore, libraryvisitbefore, libraryvisit,  itaccess, usecomp, usedevices), columns = colnames)
    newdf.replace({'5000':None, '5000.0':None}, inplace = True)
    for app in app_columns:
        newdf[app] = newdf[app].astype(int)
        newdf[app].replace({66:'not_rated', 99:'not_used'}, inplace = True)# for app in app_columns]
    newdf['Age'] = newdf['Age'].astype(int)
    newdf['CodedBefore'] = newdf['CodedBefore'].astype(int)
    newdf['CodedBefore'].replace({1:'Alot', 2:'Alittle', 3:'No', 4:'No'}, inplace = True)
    newdf.to_csv('WorkshopData/'+liblocation+'.csv')
```
